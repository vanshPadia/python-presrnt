```python
def main_menu():
    while True:
        print("\nChoose the data structure you want to interact with:")
        print("1. List")
        print("2. Dictionary")
        print("3. String")
        print("4. Tuple")
        print("5. Set")
        print("6. Exit")

        choice = input("Enter your choice (1-6): ")

        if choice == '1':
            interact_with_list()
        elif choice == '2':
            interact_with_dict()
        elif choice == '3':
            interact_with_string()
        elif choice == '4':
            interact_with_tuple()
        elif choice == '5':
            interact_with_set()
        elif choice == '6':
            print("Exiting...")
            break
        else:
            print("Invalid choice. Please enter a number from 1 to 6.")

def interact_with_list():
    my_list = []
    
    # Take input for how many items to add
    try:
        num_items = int(input("How many items do you want to add to the list? "))
    except ValueError:
        print("Invalid input. Please enter a valid number.")
        return
    
    # Loop to add items
    for i in range(num_items):
        item = input(f"Enter item {i + 1}: ")
        my_list.append(item)
    
    while True:
        print("\nOptions:")
        print("1. Add item to list")
        print("2. Remove item from list")
        print("3. Sort list")
        print("4. Print list")
        print("5. Exit")
        
        choice = input("Enter your choice (1-5): ")
        
        if choice == '1':
            item = input("Enter item to add: ")
            my_list.append(item)
            print(f"Added '{item}' to the list.")
        
        elif choice == '2':
            if not my_list:
                print("List is empty.")
            else:
                print("Current list:", my_list)
                item = input("Enter item to remove: ")
                if item in my_list:
                    my_list.remove(item)
                    print(f"Removed '{item}' from the list.")
                else:
                    print(f"'{item}' not found in the list.")
        
        elif choice == '3':
            my_list.sort()
            print("List sorted successfully.")
        
        elif choice == '4':
            if not my_list:
                print("List is empty.")
            else:
                print("Current list:", my_list)
        
        elif choice == '5':
            print("Exiting...")
            break
        
        else:
            print("Invalid choice. Please enter a number from 1 to 5.")

def interact_with_dict():
    my_dict = {}
    
    # Take input for how many key-value pairs to add initially
    try:
        num_items = int(input("How many key-value pairs do you want to add to the dictionary? "))
    except ValueError:
        print("Invalid input. Please enter a valid number.")
        return
    
    # Loop to add key-value pairs
    for i in range(num_items):
        key = input(f"Enter key {i + 1}: ")
        value = input(f"Enter value for {key}: ")
        my_dict[key] = value
    
    while True:
        print("\nOptions:")
        print("1. Add key-value pair")
        print("2. Remove key")
        print("3. Print dictionary")
        print("4. Get value by key")
        print("5. Update value by key")
        print("6. Clear dictionary")
        print("7. Exit")
        
        choice = input("Enter your choice (1-7): ")
        
        if choice == '1':
            key = input("Enter key: ")
            value = input("Enter value: ")
            my_dict[key] = value
            print(f"Added '{key}: {value}' to the dictionary.")
        
        elif choice == '2':
            key = input("Enter key to remove: ")
            if key in my_dict:
                del my_dict[key]
                print(f"Removed key '{key}' from the dictionary.")
            else:
                print(f"Key '{key}' not found in the dictionary.")
        
        elif choice == '3':
            print("Current dictionary:", my_dict)
        
        elif choice == '4':
            key = input("Enter key to get value: ")
            value = my_dict.get(key, "Key not found")
            print(f"Value for '{key}': {value}")
        
        elif choice == '5':
            key = input("Enter key to update: ")
            if key in my_dict:
                value = input(f"Enter new value for '{key}': ")
                my_dict[key] = value
                print(f"Updated '{key}' to '{value}' in the dictionary.")
            else:
                print(f"Key '{key}' not found in the dictionary.")
        
        elif choice == '6':
            my_dict.clear()
            print("Dictionary cleared.")
        
        elif choice == '7':
            print("Exiting...")
            break
        
        else:
            print("Invalid choice. Please enter a number from 1 to 7.")


def interact_with_set():
    my_set = set()
    
    # Take input for how many items to add initially
    try:
        num_items = int(input("How many items do you want to add to the set? "))
    except ValueError:
        print("Invalid input. Please enter a valid number.")
        return
    
    # Loop to add items
    for i in range(num_items):
        item = input(f"Enter item {i + 1}: ")
        my_set.add(item)
    
    while True:
        print("\nOptions:")
        print("1. Add item to set")
        print("2. Remove item from set")
        print("3. Print set")
        print("4. Check if item exists in set")
        print("5. Clear set")
        print("6. Exit")
        
        choice = input("Enter your choice (1-6): ")
        
        if choice == '1':
            item = input("Enter item to add: ")
            my_set.add(item)
            print(f"Added '{item}' to the set.")
        
        elif choice == '2':
            item = input("Enter item to remove: ")
            if item in my_set:
                my_set.remove(item)
                print(f"Removed '{item}' from the set.")
            else:
                print(f"'{item}' not found in the set.")
        
        elif choice == '3':
            print("Current set:", my_set)
        
        elif choice == '4':
            item = input("Enter item to check: ")
            if item in my_set:
                print(f"'{item}' exists in the set.")
            else:
                print(f"'{item}' does not exist in the set.")
        
        elif choice == '5':
            my_set.clear()
            print("Set cleared.")
        
        elif choice == '6':
            print("Exiting...")
            break
        
        else:
            print("Invalid choice. Please enter a number from 1 to 6.")


def interact_with_tuple():
    my_tuple = ()
    
    # Take input for how many items to add initially
    try:
        num_items = int(input("How many items do you want to add to the tuple? "))
    except ValueError:
        print("Invalid input. Please enter a valid number.")
        return
    
    # Loop to add items
    items = []
    for i in range(num_items):
        item = input(f"Enter item {i + 1}: ")
        items.append(item)
    my_tuple = tuple(items)
    
    while True:
        print("\nOptions:")
        print("1. Add item to tuple")
        print("2. Remove item from tuple")
        print("3. Print tuple")
        print("4. Get item by index")
        print("5. Find index of item")
        print("6. Count occurrences of item")
        print("7. Exit")
        
        choice = input("Enter your choice (1-7): ")
        
        if choice == '1':
            item = input("Enter item to add: ")
            my_tuple = my_tuple + (item,)
            print(f"Added '{item}' to the tuple.")
        
        elif choice == '2':
            item = input("Enter item to remove: ")
            if item in my_tuple:
                temp_list = list(my_tuple)
                temp_list.remove(item)
                my_tuple = tuple(temp_list)
                print(f"Removed '{item}' from the tuple.")
            else:
                print(f"'{item}' not found in the tuple.")
        
        elif choice == '3':
            print("Current tuple:", my_tuple)
        
        elif choice == '4':
            try:
                index = int(input("Enter index to get item: "))
                print(f"Item at index {index}: {my_tuple[index]}")
            except (ValueError, IndexError):
                print("Invalid index.")
        
        elif choice == '5':
            item = input("Enter item to find index: ")
            if item in my_tuple:
                index = my_tuple.index(item)
                print(f"Index of '{item}': {index}")
            else:
                print(f"'{item}' not found in the tuple.")
        
        elif choice == '6':
            item = input("Enter item to count occurrences: ")
            count = my_tuple.count(item)
            print(f"'{item}' occurs {count} times in the tuple.")
        
        elif choice == '7':
            print("Exiting...")
            break
        
        else:
            print("Invalid choice. Please enter a number from 1 to 7.")

def interact_with_string():
    # Take input for the initial string
    my_string = input("Enter the initial string: ")
    
    while True:
        print("\nOptions:")
        print("1. Add substring to string")
        print("2. Remove substring from string")
        print("3. Print string")
        print("4. Get character by index")
        print("5. Find index of substring")
        print("6. Count occurrences of substring")
        print("7. Get length of string")
        print("8. Exit")
        
        choice = input("Enter your choice (1-8): ")
        
        if choice == '1':
            substring = input("Enter substring to add: ")
            my_string += substring
            print(f"Added '{substring}' to the string.")
        
        elif choice == '2':
            substring = input("Enter substring to remove: ")
            if substring in my_string:
                my_string = my_string.replace(substring, '', 1)
                print(f"Removed '{substring}' from the string.")
            else:
                print(f"'{substring}' not found in the string.")
        
        elif choice == '3':
            print("Current string:", my_string)
        
        elif choice == '4':
            try:
                index = int(input("Enter index to get character: "))
                print(f"Character at index {index}: {my_string[index]}")
            except (ValueError, IndexError):
                print("Invalid index.")
        
        elif choice == '5':
            substring = input("Enter substring to find index: ")
            if substring in my_string:
                index = my_string.index(substring)
                print(f"Index of '{substring}': {index}")
            else:
                print(f"'{substring}' not found in the string.")
        
        elif choice == '6':
            substring = input("Enter substring to count occurrences: ")
            count = my_string.count(substring)
            print(f"'{substring}' occurs {count} times in the string.")
        
        elif choice == '7':
            length = len(my_string)
            print(f"Length of the string: {length}")
        
        elif choice == '8':
            print("Exiting...")
            break
        
        else:
            print("Invalid choice. Please enter a number from 1 to 8.")

main_menu()


```
